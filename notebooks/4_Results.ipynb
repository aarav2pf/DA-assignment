```python
import pandas as pd
import matplotlib.pyplot as plt

#Load saved Apriori and FP-Growth rule files
apriori_rules = pd.read_csv("../outputs/rules/apriori_rules.csv")
fp_rules = pd.read_csv("../outputs/rules/fpgrowth_rules.csv")

#Keep only the top 5 rules (sorted by lift)
apriori_top = apriori_rules.sort_values("lift", ascending=False).head(5).reset_index(drop=True)
fp_top = fp_rules.sort_values("lift", ascending=False).head(5).reset_index(drop=True)

#Display tables of top rules
print("Top 5 Apriori Rules\n")
display(apriori_top)

print("\nTop 5 FP-Growth Rules\n")
display(fp_top)

#Plot top 5 Apriori rules by lift
plt.figure(figsize=(8, 4))
plt.barh(apriori_top['antecedents'] + "  " + apriori_top['consequents'],
         apriori_top['lift'])
plt.xlabel("Lift")
plt.ylabel("Rule")
plt.title("Top 5 Apriori Rules by Lift")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

#Plot top 5 FP-Growth rules by lift
plt.figure(figsize=(8, 4))
plt.barh(fp_top['antecedents'] + "  " + fp_top['consequents'],
         fp_top['lift'], color="lightcoral")
plt.xlabel("Lift")
plt.ylabel("Rule")
plt.title("Top 5 FP-Growth Rules by Lift")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
